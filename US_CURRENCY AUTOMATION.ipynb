```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
url = 'https://www.cbp.gov/sites/default/files/assets/documents/2023-Jul/Daily_Currency_Notice_07-24-2023.xlsx'
response = requests.get(url)
```

```python
if response.status_code == 200:
    with open('data.xlsx', 'wb') as f:
        f.write(response.content)
else:
    print("Failed to download the file")
```

```python
df = pd.read_excel('data.xlsx', sheet_name='Daily Currency')
df
```

```python
df = df[df.columns[1:]]
df = df.dropna()
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)
df
```

```python
file_path = r'C:\Users\MOHAN_E0313\Downloads\CDR_CURRENCY_CONVERSION_CUSTOMS_072423.xlsx'
df2 = pd.read_excel(file_path, sheet_name = 'Sheet1' )

df2
```

```python
date_str = url.split('/')[-1].split('_')[-1].split('.')[0]
date = pd.to_datetime(date_str, format='%m-%d-%Y')
```

```python
date = date.strftime('%d-%b-%y')
date
```

```python
merged_table = pd.merge(df2, df, left_on='COUNTRY_CODE FROM', right_on='ISO Code', suffixes=('_df2', '_df'))

columns_to_drop = ['ISO Code', 'Country/Union', 'Rate_df2', 'IND', 'CUR Code']
 #Drop the additional C1 column from the merged DataFrame
merged_table.drop(columns_to_drop, axis=1, inplace=True)


# Define the new column name for the column you want to change
new_column_name = 'Rate'

# Assign the new column name to the specific column
merged_table.rename(columns={'Rate_df': new_column_name}, inplace=True)


# Define the desired order of columns
desired_order = ['COUNTRY_CODE', 'COUNTRY_CODE FROM', 'CURRENCY_FROM', 'CURRENCY_TO','Rate','RATE_TYPE','EFFECTIVE_FROM_DATE','EFFECTIVE_TO_DATE','STATUS']

# Rearrange the columns using reindex()
merged_table = merged_table.reindex(columns=desired_order)

merged_table
```

```python
# Replace string representations with actual NaN values
Final_table = merged_table.replace({'NaN': np.nan, 'N/A': np.nan})

# Now you can use fillna() to replace NaNs with 'Y'
Final_table.fillna('Y', inplace=True)

Final_table.info()
```

```python
Final_table['EFFECTIVE_FROM_DATE'] = date
Final_table['EFFECTIVE_TO_DATE'] = date
```
